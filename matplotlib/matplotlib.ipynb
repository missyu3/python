```python
# matplotlib
# データの可視化ツール　図を簡単に描画できる
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
x = [1,2,3,4,5]
y = np.array(x) ** 2
plt.plot(x,y,"g--",linewidth = 5)
```

```python
x = np.linspace(-10,10,100)
z = 1 / (1 + np.exp(-x))
plt.plot(x,z)
```

```python
df = pd.DataFrame({"a":[1,2,3,4,5,6],
                   "b":[10,15,40,100,200,400]})
```

```python
plt.plot("a","b",data = df)
```

```python
plt.plot(x,y,
         color="red",
        lw = 1,
        ls = "--",
         marker="o",
         markersize=11,
         markerfacecolor = "yellow",
         markeredgecolor = "blue",
         markeredgewidth = 3,
         alpha = 0.3,
         label = "sigmoid"
        )
plt.xlabel("x")
plt.ylabel("y")
plt.title("sigmoid(x)")
plt.legend(loc = 0)
plt.xticks([1,2,3],rotation=20)
plt.yticks([5,10,25],rotation=20)
```
